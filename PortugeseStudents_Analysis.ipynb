```python
import pandas as pd

'''
This dataset is a student performance dataset. 
It combines demographics, family background, school-related features, and lifestyle habits, 
  then ties them to school grades (G1, G2, G3).
'''
StudentData = pd.read_csv ('student-por.csv', sep=';')

# display the first few rows of the dataset
firstRows = StudentData.head()
print (firstRows)
```

```python
#number of rows and columns
print (StudentData.shape)
```

```python
#list ofcolumn names
print (StudentData.columns)
```

```python
# count missing values in each column
print(StudentData.isnull().sum())
```

```python
# check if there are missing values anywhere in the dataset
print(StudentData.isnull().values.any())
```

```python
#to check the datatype in each column
print(StudentData.dtypes)
```

```python
# basic statistics
print(StudentData.describe())
```

```python
#including categorical columns
print(StudentData.describe(include='all'))
```

```python
# mean grades grouped by school
print(StudentData.groupby("school")[["G1", "G2", "G3"]].mean())

#GP - Gabriel Pereira students perform better than MS - Mousinho da Silveira students
```

```python
#mean grades of grades grouped by sex
print(StudentData.groupby("sex")[["G1", "G2", "G3"]].mean())

#females perform slightly better than males
```

```python
#Do students who study more hours get better grades?
StudentData.groupby("studytime")["G3"].mean() # studnets with more study time had higher grades(3,4)
```

```python
#does parent educational level affect grades?
print(StudentData.groupby(["Medu", "Fedu"])[["G1", "G2", "G3"]].mean()) 
#higher parental education is associated with slightly higher grades.
```

```python
#Does student attendance affect grade?
print(StudentData.groupby('absences')[['G1', 'G2', 'G3']].mean())
```

```python
# Correlation between absences and final grade (G3)
correlation = StudentData["absences"].corr(StudentData["G3"])
print(correlation)

'''
Absences do not strongly predict student performance. While some students with very high absences still performed well, 
the overall correlation is close to zero, 
suggesting that factors beyond attendance (like parental education, study time, or support) might be more influential.
'''
```

```python
#Does paid classes help with students grade in Portugeese?
print(StudentData.groupby('paid')[["G1", "G2", "G3"]].mean())

#Students who didnâ€™t take paid classes performed slightly better on all three grade metrics.
#This suggests that extra classes may not necessarily contribute to academic performance
```

```python
#Does travel time affect student grade?
print(StudentData.groupby('traveltime')[['G1', 'G2', 'G3']].mean())

#Analysis suggest that students that spend less time commuting to school (Groups 1 and 2) perform excellently better to other groups
#Group 3 and 4 who spend over 1 hour to get to school.
```

```python
import matplotlib.pyplot as plt 
```

```python
# Compute mean grades across all students
avg_grades = StudentData[["G1", "G2", "G3"]].mean()

plt.plot(avg_grades.index, avg_grades.values, marker='o', linestyle='-', color='b')
plt.title("Average Grade Progression (G1 â†’ G3)")
plt.xlabel("Grade Period")
plt.ylabel("Average Grade")
plt.grid(True)
plt.show()
```

```python
#plot of free time against mean of final grade

import matplotlib.pyplot as plt

avg_freetime = StudentData.groupby('freetime')['G3'].mean()

plt.plot(avg_freetime.index, avg_freetime.values, marker='o', linestyle='-', color='r')
plt.title("Average Final Grade by Free Time Level")
plt.xlabel("Free Time Level (1 = Very Low, 5 = Very High)")
plt.ylabel("Average Final Grade (G3)")
plt.show()
```

```python
#plot of romantic relationship affects students grade in Portugeese

import matplotlib.pyplot as plt

studentRomantic = StudentData.groupby('romantic')['G3'].mean()

plt.bar(studentRomantic.index, studentRomantic.values, color="skyblue", edgecolor="black")
plt.title('Effect of Relationship on students average performance')
plt.xlabel('Relationship Status: yes or no')
plt.ylabel("Average Final Grade (G3)")
plt.show()
```

```python
#plot of how family educational support affects students grade in Portugeese

import matplotlib.pyplot as plt

studentRomantic = StudentData.groupby('famsup')['G3'].mean()

plt.bar(studentRomantic.index, studentRomantic.values, color="Green", edgecolor="black")
plt.title('Effect of Family educational support on students average performance')
plt.xlabel('Family educational support: yes or no')
plt.xticks(rotation=0)
plt.ylabel("Average Final Grade (G3)")
plt.show()
```

```python
#Distribution of final grade with histogram

StudentData["G3"].plot(kind="hist", bins=15, color="yellow", edgecolor="black")
plt.title("Distribution of Final Grades (G3)")
plt.xlabel("Final Grade")
plt.ylabel("Frequency")
plt.show()
```

```python
#final grade based on study time

StudentData.groupby("studytime")["G3"].mean().plot(
    kind="bar", color="orange", edgecolor="black"
)
plt.title("Average Final Grade by Study Time")
plt.xlabel("Study Time Group")
plt.ylabel("Average Final Grade (G3)")
plt.show()
```

```python
#Plot of the number fathers in each category

import matplotlib.pyplot as plt

# Count how many students fall into each Fjob category
father_jobs = StudentData['Fjob'].value_counts()

# Plot pie chart
plt.pie(
    father_jobs.values, 
    labels=father_jobs.index, 
    autopct='%1.1f%%',  # show percentages with 1 decimal
    startangle=90,      # rotate start 
    colors=plt.cm.Pastel1.colors  # soft color palette
)

plt.title("Distribution of Fathers' Jobs")
plt.show()
```
